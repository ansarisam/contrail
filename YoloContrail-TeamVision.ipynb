```python
import glob
images_list = glob.glob("/content/drive/MyDrive/yolov3/data/obj/images/*.png")
print(images_list)
```

```python
file_train = open("/content/drive/MyDrive/yolov3/data/train.txt", "w")   
file_test = open("/content/drive/MyDrive/yolov3/data/test.txt", "w")
```

```python
dataset_path = '/content/drive/MyDrive/yolov3/data/obj/images'
```

```python
import os
counter = 1  
index_test = round(100 / 10)  
for pathAndFilename in images_list:  
    title, ext = os.path.splitext(os.path.basename(pathAndFilename))

    if counter == index_test+1:
        counter = 1
        file_test.write(dataset_path + "/" + title + '.png' + "\n")
    else:
        file_train.write(dataset_path + "/" + title + '.png' + "\n")
        counter = counter + 1
```

```python
import glob
import os
import re

txt_file_paths = glob.glob(r"/content/drive/MyDrive/yolov3/data/obj/images/*.txt")
for i, file_path in enumerate(txt_file_paths):
    # get image size
    with open(file_path, "r") as f_o:
        lines = f_o.readlines()

        text_converted = []
        for line in lines:
            print(line)
            numbers = re.findall("[0-9.]+", line)
            print(numbers)
            if numbers:

                # Define coordinates
                text = "{} {} {} {} {}".format(0, numbers[1], numbers[2], numbers[3], numbers[4])
                text_converted.append(text)
                print(i, file_path)
                print(text)
        # Write file
        with open(file_path, 'w') as fp:
            for item in text_converted:
                fp.writelines("%s\n" % item)
```
